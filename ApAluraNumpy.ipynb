DIFERENCIA EN TIEMPO AL USAR ARRAYS

```python
import numpy as np
import time

# crea una lista con 1000000 elementos
lista = list(range(1000000))

# convierte la lista en un array Numpy
array = np.array(lista)

# comienza a medir el tiempo para la operación con la lista
start_time = time.time()

# realiza la operación de elevar al cuadrado cada elemento de la lista
lista_cuadrado = [i**2 for i in lista]

# detiene el cronómetro
tiempo_lista = time.time() - start_time

# comienza a medir el tiempo para la operación con el array
start_time = time.time()

# realiza la operación de elevar al cuadrado cada elemento del array
array_cuadrado = array**2

# detiene el cronómetro
tiempo_array = time.time() - start_time

print("Tiempo de la operación con la lista: ", tiempo_lista)
print("Tiempo de la operación con el array: ", tiempo_array)
```

```python
# Generar etiquetas de año aproximadas
años = [str(2013 + i // 12) for i in range(len(fechas))]

# Como los datos cubren un poco más de 7 años, ajustamos los últimos años para reflejar eso
# Por simplicidad, mostraremos el año al comienzo de cada año en el eje x
años_unicos = []
fechas_años_unicos = []
for i in range(len(fechas)):
    if i % 12 == 0:
        años_unicos.append(str(2013 + i // 12))
        fechas_años_unicos.append(fechas[i])

plt.figure(figsize=(12, 6)) # Ajustar el tamaño del gráfico para mejor visualización
plt.plot(fechas, Moscu, label='Moscu')
plt.plot(fechas, Kaliningrado, label='Kaliningrado')
plt.plot(fechas, Petersburgo, label='Petersburgo')
plt.plot(fechas, Krasnodar, label='Krasnodar')
plt.plot(fechas, Ekaterimburgo, label='Ekaterimburgo')

plt.xticks(fechas_años_unicos, años_unicos) # Establecer las etiquetas del eje X
plt.xlabel('Año') # Etiqueta del eje X
plt.ylabel('Precio') # Etiqueta del eje Y
plt.title('Precios de Manzanas por Ciudad a lo Largo del Tiempo') # Título del gráfico

plt.legend()
plt.show()
```
